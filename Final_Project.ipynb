```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Berlin_crimes.csv')
```

```python
df.head()
```

### Overview

```python
#rename Agg_assault for better understanding
berlin = df.rename(columns={'Agg_assault': 'Aggravated_assault'})
```

```python
#convert all columns to lower case

berlin.columns = berlin.columns.str.lower()
```

```python
#dropping local column because I'm not sure what this means and code becuase its not needed
berlin = berlin.drop('local', axis=1)
berlin = berlin.drop('code',axis=1)
```

```python
berlin.head(2)
```

```python
#unique number of years
berlin['year'].unique()
```

```python
#unique number of district
berlin['district'].unique()
```

```python
#unique number of locations
berlin['location'].unique()
```

```python
#grouping locations by district
pd.options.display.max_rows = None #views maximum number of rows
grouped = berlin.groupby('district')['location']
value_counts = grouped.value_counts()
print(value_counts)
```

```python
# Total number of crimes per the crimes

# identify nominal columns with less than 10000 unique values
nominal_cols = []
for col in berlin.columns:
    if berlin[col].nunique() <= 10000 and col not in ['district', 'location']: # excluding categorical columns
        nominal_cols.append(col)

# calculate the sum of each nominal column
column_sums = {}
for col in nominal_cols:
    column_sum = berlin[col].sum()
    column_sums[col] = column_sum

# print the results
print(column_sums)
```

```python
# grouping districts by locations

grouped = berlin.groupby('district')['location']
value_counts = grouped.value_counts()
print(value_counts)
```

```python
# group the crimes committed by year
grouped_df = berlin.groupby('year').sum()

# print the resulting dataframe
print(grouped_df)
```

```python
# group the crimes committed by year
grouped_df = berlin.groupby('location').sum()

# print the resulting dataframe
print(grouped_df)
```
